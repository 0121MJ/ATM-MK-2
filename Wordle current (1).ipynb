```python
user_entry = 'sauce' #to start with
word_otd = 'spire'
prev_guesses = []

for main_iter in range (1000):

    prev_guesses.append(user_entry)
    
    with open('wordle-answers-alphabetical.txt') as f:
        lines = f.readlines()
    all_words = [i.replace("\n","") for i in lines]

    all_words_in_characters = []

    def extract_letters(word):
        book = ["","","","",""]
        for i in range(0,len(word)):
            book[i] = word[i]
        return book

    for i in all_words:
        all_words_in_characters.append(extract_letters(i))

    # this produces the word a "user" enters as a guess of the word otd

    user_entry_in_chars = []

    user_entry_in_chars = extract_letters(user_entry)
    print("User entry:")
    print(user_entry_in_chars)


    # this produces the word of the day that is to be guessed

    word_otd_in_chars = []

    word_otd_in_chars = extract_letters(word_otd)
    print("Word of the day:")
    print(word_otd_in_chars)


    # this identifies which letters are green
    foundchars = ["#","#","#","#","#"]
    foundchar_loc = []

    for i in range(len(user_entry_in_chars)):
        if user_entry_in_chars[i] == word_otd_in_chars[i]:
            foundchars[i] = str(user_entry_in_chars[i])
            foundchar_loc.append(i)
    print("Greens:")
    print(foundchars)


    # After the greens are identified, this identifies the remaining letters the user entered
    potential_yellows = []
    for i in range(0,5):
        if i in foundchar_loc:
            continue
        potential_yellows.append(user_entry_in_chars[i])

    print("Potential yellows:")
    print(potential_yellows)


    # this determines which letters the user still needs to find
    letters_not_found_by_user = []
    for i in range(0,5):
        if i in foundchar_loc:
            continue
        letters_not_found_by_user.append(word_otd[i])

    print("Letters user has left to find:")
    print(letters_not_found_by_user)


    # this shows which letters the user guessed which are yellows
    yellows = []
    for i in range(len(potential_yellows)):
        if potential_yellows[i] in letters_not_found_by_user:
            yellows.append(potential_yellows[i])

    print("Actual yellows:")
    print(yellows)


    # This filters through the list to select only words which have the same letters as the 
    remaining_words0 = all_words_in_characters
    remaining_words1 = []
    remaining_words2 = []
    remaining_words3 = []
    remaining_words4 = []
    remaining_words5 = []

    for letter_no in range(len(foundchars)):
        if foundchars[letter_no] == '#':
            locals()["remaining_words"+str(letter_no+1)] = locals()["remaining_words"+str(letter_no)]
            continue
        else:
            for word_no in range(len(locals()["remaining_words"+str(letter_no)])):
                if locals()["remaining_words"+str(letter_no)][word_no][letter_no] == foundchars[letter_no]:
                    locals()["remaining_words"+str(letter_no+1)].append(locals()["remaining_words"+str(letter_no)][word_no])
            continue

    # this filters out all the words that contain only the yellow letters
    final_remaining_words = []

    if yellows != []:

        for i in range(len(yellows)):
            for j in remaining_words4:
                if all(elem in j for elem in yellows):
                    final_remaining_words.append(j)
    else: final_remaining_words = remaining_words4

    #Given how common each letter is, this detects which word is best
    with open('wordle-answers-alphabetical.txt') as f:
        lines = f.readlines()
    all_words = [i.replace("\n","") for i in lines]

    # These are the frequencies of letters from wikipedia
    letter_scores = {'a':7.8,'b':2,'c':4,'d':3.8,'e':11,
                     'f':1.4,'g':3,'h':2.3,'i':8.2,'j':0.21,
                     'k':2.5,'l':5.3,'m':2.7,'n':7.2,'o':6.1,
                     'p':2.8,'q':0.24,'r':7.3,'s':8.7,'t':6.7,
                     'u':3.3,'v':1,'w':0.91,'x':0.27,'y':1.6,
                     'z':0.44,'#':0}
    #This algorithm removes letters that appear more than once 
    A = []
    B = []
    C = []
    D = []
    E = []
    counter = 0
    for word in final_remaining_words:
        for i in range(5):
            if i == 0:
                A = word[i]
            if i == 1:
                B = word[i]
            if i == 2:
                C = word[i]
            if i == 3:
                D = word[i]
            if i == 4:
                E = word[i]
        if A == B or A == C or A == D or A == E:
            word[0] = "#"
        if B == C or B == D or B == E:
            word[1] = "#"
        if C == D or C == E:
            word[2] = "#"
        if D == E:
            word[3] = "#"

    # this calculates the scores of the remaining words given how common they are
    words_scores = []
    for i in range(len(final_remaining_words)):
        words_scores.append(0)
    words_scores

    counter = 0

    for word in final_remaining_words:
        for j in range(4):
            words_scores[counter] += letter_scores[word[j]]
        counter += 1

    words_scores
    final_remaining_words1 = []
    counter = 0

    for i in final_remaining_words:
        final_remaining_words1.append(i[0] + i[1] + i[2] + i[3] + i[4])
        
    # this makes a dictionary joining the remaining possible words by their scores
    dictionary = {final_remaining_words1[i]: words_scores[i] for i in range(len(final_remaining_words1))}
    
    # this removes the previous guesses from the candidates 
    for i in prev_guesses:
        try:
            del dictionary[i]
        except:
            continue
    
    print(dictionary)

    print(f"\n\nScript guess no. {main_iter+1} is {max(dictionary, key=dictionary.get)}\n")
    
    if word_otd == max(dictionary, key=dictionary.get):
          print(f"\nWord found succesfully at attempt {main_iter+1}")
          break
    user_entry = max(dictionary, key=dictionary.get)
    
    if main_iter == 1000:
        print("Unable to find word in 1000 tries")
```
