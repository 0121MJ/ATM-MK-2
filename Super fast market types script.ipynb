```python
# This script is significantly faster than the others, but provides less of a guarantee of accuracy due to differences
# in the naming conventions used by the site it relies on (sharecast.com) and the LSE

import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

df = pd.read_excel("input.xlsx")
cols = df["Company Name"].tolist()
cols = [i.upper().replace("PLC","").strip() for i in cols]
print("\nUser firms:")
for i in cols:
    print(i)
      

market_types = {"FTSE100":"FTSE_100",
                "FTSE250":"FTSE_250",
                "AIM":"FTSE_AIM_All-Share",
                "SmallCap":"FTSE_Small_Cap"}
FTSE100 = []
FTSE250 = []
AIM = []
SmallCap = []
      
for market_type in market_types:

    print(f"\nNow collecting {market_type} index")
    response = requests.get(f"https://www.sharecast.com/index/{market_types[market_type]}")
    soup = BeautifulSoup(response.text, 'html.parser')
    constituents = soup.find_all("a", {"target":"_top"})
    constituents = [i.get_text().upper().replace('(DI)','').replace('(CDI)','').replace('GBP','').strip() for i in constituents]
    if market_type == "FTSE100":
        for i in constituents:
            FTSE100.append(i)
    if market_type == "FTSE250":
        for i in constituents:
            FTSE250.append(i)
    if market_type == "AIM":
        for i in constituents:
            AIM.append(i)
    if market_type == "SmallCap":
        for i in constituents:
            SmallCap.append(i)

print(f"\nNow searching user input in indexes")   

output = []

for i in cols:
    if i in FTSE100:
        output.append("FTSE 100")
        continue
    if i in FTSE250:
        output.append("FTSE 250")
        continue
    if i in AIM:
        output.append("AIM")
        continue
    if i in SmallCap:
        output.append("SmallCap")
        continue
    output.append("Not found")
    print(f"Identified {i}")

df = pd.DataFrame(output)
df.to_excel("output_file.xlsx")

print("\nScript complete")
```
